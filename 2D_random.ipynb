```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.zeros(2)
b = np.zeros((1000,2))
for i in range(1,1000):
    theta = np.random.random()*2*np.pi-np.pi
    a += np.array([np.cos(theta),np.sin(theta)])
    b[i] = a
plt.plot(b[:,0],b[:,1])
```

```python
a = np.zeros(2)
b = np.zeros((1000,2))
for i in range(1,1000):
    a += np.random.randint(0,2,size=2)-1/2
    b[i] = a
plt.plot(b[:,0],b[:,1])
```

```python
a = np.zeros(2)
b = np.zeros((1000,2))
for i in range(1,1000):
    a += np.random.randint(0,2,size=2)*2-1
    b[i] = a
plt.plot(b[:,0],b[:,1])
```
